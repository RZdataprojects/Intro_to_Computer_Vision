# Gaussion Filters
Bluring images

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('/content/drive/MyDrive/media/Saturn_during_Equinox.jpg')
cv2_imshow(img)
```

adding noise

```python
noise = np.random.normal(loc = img.mean(), scale = img.std(), size = img.shape)
img_wn = img + noise
cv2_imshow(img_wn)
```

using a Gaussion Filter to blur the image

```python
from scipy.ndimage import gaussian_filter

result = gaussian_filter(img, sigma=5)
cv2_imshow(result)
```
