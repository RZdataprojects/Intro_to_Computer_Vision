```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#input image
img = cv.imread('/content/drive/MyDrive/מדיה/apple.jpg', 0) #flag=0 - grayscale
```

```python
#show image
cv2_imshow(img)
```

```python
#impge resolution H, W, number of color planes - in this case non as the image is black and white
print(img.shape)
```

```python
#get the intensity value of a single pixel
img[500,100]
```

```python
#ploting the intensity of a single row of pixels
plt.plot(img[200,:])
```

```python
#crop an image
cropped_image = img[200:500, 500:800]
cv2_imshow(cropped_image)
```

```python
#note that the cropped image includes the end rows and columns
print(cropped_image.shape)
```
